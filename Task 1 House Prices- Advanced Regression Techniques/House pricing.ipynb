```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
train_data=pd.read_csv('train.csv')
train_data
```

```python
train_data = train_data[train_data["GrLivArea"] < 4000]
```

```python
test_data=pd.read_csv('test.csv')
test_data
```

```python
train_data.info()
```

```python
test_data.info()
```

```python
def fix_missing_values(df):
    df = df.copy()

    for col in df.select_dtypes(include=["number"]).columns:
            df[col] = df[col].fillna(df[col].median())
    for col in df.select_dtypes(include=["object"]).columns:
        df[col] = df[col].fillna("None")

    return df
```

```python
train_data = fix_missing_values(train_data)
```

```python
test_data = fix_missing_values(test_data)
```

```python
categorical_features = train_data.select_dtypes(include=["object"]).columns
for col in categorical_features:
    le = LabelEncoder()
    train_data[col]= le.fit_transform(train_data[col].astype(str))

    test_data[col] = test_data[col].map(lambda x: le.classes_.tolist().index(x) if x in le.classes_ else -1)
```

```python
train_data["Total_SquareFeet"]=train_data[["TotalBsmtSF","1stFlrSF","2ndFlrSF"]].apply(pd.to_numeric, errors="coerce").sum(axis=1)
test_data["Total_SquareFeet"]=train_data[["TotalBsmtSF","1stFlrSF","2ndFlrSF"]].apply(pd.to_numeric, errors="coerce").sum(axis=1)
```

```python
features = [col for col in train_data.columns if col not in ["Id", "SalePrice"]]
```

```python
y= np.log1p(train_data["SalePrice"])
```

```python
x = train_data[features]
x_test = test_data[features]
```

```python
x.describe()
```

```python
x_train, x_validation, y_train, y_validation = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
XGB_model = xgb.XGBRegressor(n_estimators=1000, learning_rate=0.05, random_state=42)
try:
    XGB_model.fit(x_train, y_train, early_stopping_rounds=50, eval_set=[(x_validation, y_validation)], verbose=False)
except TypeError:
    XGB_model.fit(x_train, y_train, eval_set=[(x_validation, y_validation)], verbose=False)
```

```python
RF_model = RandomForestRegressor(n_estimators=100, random_state=42)
RF_model.fit(x_train, y_train)
```

```python
Random_pred = np.expm1(RF_model.predict(x_test))
```

```python
XGB_pred = np.expm1(XGB_model.predict(x_test))
```

```python
prediction=(Random_pred + XGB_pred) / 2
```

```python
submission=pd.DataFrame({'Id':test_data['Id'],'SalePrice':prediction})
submission.to_csv('submission.csv',index=False)
print("submission successfully")
```
